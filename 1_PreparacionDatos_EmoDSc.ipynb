Eliminacin de silencios

```python
import pandas as pd
from pydub import AudioSegment
from pydub.utils import which
```

```python
AudioSegment.converter = which("ffmpeg")

EmoDSc_Balanceado = pd.read_csv('CSVs\\EmoDSc_Balanceado.csv')

location = 'EmoDSc_Silenciados\\'

def detect_leading_silence(dls_audio):
    dls_chunk_size = 10
    dls_silence_threshold = -50.0

    trim_ms = 0
    dls_audio_length = len(dls_audio)

    while trim_ms < dls_audio_length:
        chunk_dBFS = dls_audio[trim_ms:trim_ms+dls_chunk_size].dBFS
        if chunk_dBFS > dls_silence_threshold:
            break
        trim_ms += dls_chunk_size

    return trim_ms

name = 0
errores = 0
file_data = []

for Dir, Emo in zip(EmoDSc_Balanceado.Path, EmoDSc_Balanceado.Emotions):
    file_name = f'{location}{name}.wav'

    try:
        audio = AudioSegment.from_file(Dir, 'wav')
    except:
        audio = AudioSegment.from_file(Dir)

    start_trim = detect_leading_silence(audio)
    end_trim = detect_leading_silence(audio.reverse())
    
    duration = len(audio)
    new_audio = audio[start_trim:duration-end_trim]

    if len(new_audio) > 0:
        new_audio.export(file_name, format='wav')
    else:
        audio.export(file_name, format='wav')
        errores += 1
        print(f'{Dir} es un audio totalmente silencioso')

    file_data.append((Emo, file_name))

    print(f'Audios procesados: {name}, Errores encontrados: {errores}')
    name += 1

EmoDSc_Silenciados = pd.DataFrame(file_data, columns=['Emotions', 'Path'])
EmoDSc_Silenciados.to_csv("CSVs\\EmoDSc_Silenciados.csv", index=False)
```

Aumento de datos

```python
import pandas as pd
import numpy as np
import librosa
import librosa.display
import soundfile as sf
```

```python
EmoDSc_Silenciados = pd.read_csv('CSVs\\EmoDSc_Silenciados.csv')
location = 'EmoDSc_Aumentado\\'

def noise(n_data):
    noise_amp = 0.04*np.random.uniform()*np.amax(n_data)
    return n_data + noise_amp*np.random.normal(size=n_data.shape[0])

def stretch(s_data):
    s_rate = 0.8
    return librosa.effects.time_stretch(s_data, rate=s_rate)

def pitch(p_data, p_sampling_rate):
    p_pitch_factor=0.7
    return librosa.effects.pitch_shift(p_data, sr=p_sampling_rate, n_steps=p_pitch_factor)

name = 0
file_data = []

for Dir, Emo in zip(EmoDSc_Silenciados.Path, EmoDSc_Silenciados.Emotions):
    # Original
    file_name = f'{location}{name}.wav'
    print(f'Audios procesados: {name}')

    data, sample_rate = librosa.load(Dir)
    sf.write(file_name, data, sample_rate)

    file_data.append((Emo, file_name))
    name += 1

    # Ruido
    file_name = f'{location}{name}.wav'
    print(f'Audios procesados: {name}')

    data_ruido = noise(data)
    sf.write(file_name, data_ruido, sample_rate)

    file_data.append((Emo, file_name))
    name += 1 

    # Estiramiento
    file_name = f'{location}{name}.wav'
    print(f'Audios procesados: {name}')

    data_estiramiento = stretch(data)
    sf.write(file_name, data_estiramiento, sample_rate)

    file_data.append((Emo, file_name))
    name += 1 

    # Variacin de tono
    file_name = f'{location}{name}.wav'
    print(f'Audios procesados: {name}')

    data_tono = pitch(data, sample_rate)
    sf.write(file_name, data_tono, sample_rate)

    file_data.append((Emo, file_name))
    name += 1 

EmoDSc_Aumentado = pd.DataFrame(file_data, columns=['Emotions', 'Path'])
EmoDSc_Aumentado.to_csv("CSVs\\EmoDSc_Aumentado.csv", index=False)
```

```python
#FUNCIONES NECESARIAS:

def create_waveplot(data, sr, e, t):
    plt.figure(figsize=(10, 3))
    #plt.title('Audio signal {}'.format(e), size=15)
    plt.title('Seal de audio {}'.format(e), size=15)
    librosa.display.waveshow(data, sr=sr)
    #plt.xlabel('Time (s)')  # Etiqueta para el eje x
    plt.xlabel('Tiempo (s)')  # Etiqueta para el eje x
    #plt.ylabel('Amplitude')  # Etiqueta para el eje y
    plt.ylabel('Amplitud')  # Etiqueta para el eje y
    plt.tight_layout()
    plt.savefig('.\\Figuras\\waveplot_{}.eps'.format(t), format='eps', dpi=1200)
    plt.show()

def create_spectrogram(data, sr, e, t):
    # stft function converts the data into short term fourier transform
    X = librosa.stft(data)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(12, 3))
    #plt.title('Spectrogram of the audio signal {}'.format(e), size=15)
    plt.title('Espectrograma de la seal de audio {}'.format(e), size=15)
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz', cmap='inferno')   
    #librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
    #plt.colorbar(label='Amplitude (dB)')  # Etiqueta para la barra de colores
    plt.colorbar(label='Amplitud (dB)')  # Etiqueta para la barra de colores
    #plt.xlabel('Time (s)')  # Etiqueta para el eje x
    plt.xlabel('Tiempo (s)')  # Etiqueta para el eje x
    #plt.ylabel('Frequency (Hz)')  # Etiqueta para el eje y
    plt.ylabel('Frecuencia (Hz)')  # Etiqueta para el eje y
    plt.tight_layout()
    plt.savefig('.\\Figuras\\spectrogram_{}.eps'.format(t), format='eps', dpi=120)
```
